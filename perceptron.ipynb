```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input = np.linspace(-10, 10, 100)
```

```python
def sigmoid(X):
    val = 1/(1+np.exp(-X))
    return val
```

```python
output = sigmoid(input)
```
